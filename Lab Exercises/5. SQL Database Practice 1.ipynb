The `ibm_db` package is not native to Python. If unable to install package, this lab practice can be done on IBM Watson Studio.

```python
import ibm_db
```

The following credentials are need to connect to DB2:

Host DNS name or IP address (**"hostname"**)

User ID (**"username"**)

User Password (**"password"**)

Driver Name (Defined in this lab)

Database name (**"db"**)

Host port (**"port"**)

Connection protocol (**PROTOCOL**)

```python
#Replace the placeholder values with your actual Db2 hostname, username, and password:
dsn_hostname = "dashdb-txn-sbox-yp-dal09-10.services.dal.bluemix.net" # "hostname" or HOSTNAME
dsn_uid = "Hidden" # "username" or UID
dsn_pwd = "Hidden" # "password" or PWD

dsn_driver = "{IBM DB2 ODBC DRIVER}"
dsn_database = "BLUDB" # "db" or DATABASE
dsn_port = "50000" # "port" or PORT
dsn_protocol = "TCPIP" # PROTOCOL=TCPIP
```

Build the dsn connection string using the credentials above

```python
#Create the dsn connection string
dsn = (
    "DRIVER={0};"
    "DATABASE={1};"
    "HOSTNAME={2};"
    "PORT={3};"
    "PROTOCOL={4};"
    "UID={5};"
    "PWD={6};").format(dsn_driver, dsn_database, dsn_hostname, dsn_port, dsn_protocol, dsn_uid, dsn_pwd)

#Create database connection
try:
    conn = ibm_db.connect(dsn, "", "")
    print ("Connected to database!")

except:
    print ("Unable to connect: ", ibm_db.conn_errormsg() )
```

Retrieve Metadata for the Database Server

```python
server = ibm_db.server_info(conn)

print ("DBMS_NAME: ", server.DBMS_NAME)
print ("DBMS_VER:  ", server.DBMS_VER)
print ("DB_NAME:   ", server.DB_NAME)
```

Retrieve Metadata for the Database Client / Driver

```python
client = ibm_db.client_info(conn)

print ("DRIVER_NAME:          ", client.DRIVER_NAME) 
print ("DRIVER_VER:           ", client.DRIVER_VER)
print ("DATA_SOURCE_NAME:     ", client.DATA_SOURCE_NAME)
print ("DRIVER_ODBC_VER:      ", client.DRIVER_ODBC_VER)
print ("ODBC_VER:             ", client.ODBC_VER)
print ("ODBC_SQL_CONFORMANCE: ", client.ODBC_SQL_CONFORMANCE)
print ("APPL_CODEPAGE:        ", client.APPL_CODEPAGE)
print ("CONN_CODEPAGE:        ", client.CONN_CODEPAGE)
```

Finally, close the connection

```python
ibm_db.close(conn)
```
