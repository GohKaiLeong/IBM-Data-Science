The participants had three options for each topic: 'Very Interested', 'Somewhat interested', and 'Not interested'. 2,233 respondents completed the survey.

Question 1: 

Use the pandas `read_csv` method to read the csv file into a pandas dataframe

Use the `index_col` parameter in order to load the first column as the index of the dataframe

```python
import pandas as pd
```

```python
link = 'https://cocl.us/datascience_survey_data'
survey = pd.read_csv(link, index_col=0)
survey
```

Question 2:

Use the artist layer of Matplotlib to replicate the bar chart below to visualize the percentage of the respondents' interest in the different data science topics surveyed

```python
import numpy as np
%matplotlib inline 
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# convert to percentage
sur_per = (100*survey/2233).round(2)

# sort by descending order based on "very interested"
sur_per_sorted = sur_per.sort_values(by = ['Very interested'], ascending = False)
sur_per_sorted
```

```python
# Artist layer bar chart
ax = sur_per_sorted.plot(kind = 'bar',
                figsize = (20,8),
                width = 0.8,
                color = ('#5cb85c', '#5bc0de', '#d9534f'),
                fontsize = 14)
ax.set_title("Percentage of Respondents' Interest in Data Science Areas", fontsize = 16)
ax.legend(fontsize = 14, loc = 'upper right')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

# Adding the percentage values    
for p in ax.patches:
    ax.annotate("{} %".format(p.get_height()),
                xy=(p.get_x(), p.get_height()+1),fontsize=11)
```
